```python
# Preprocessing
# Separate features and target variable
```

```python
features = df.drop("Exited", axis=1)
```

```python
features
```

```python
target = df['Exited']
```

```python
target
```

```python
# Dropping unnessary columns
features = features.drop(["RowNumber", "CustomerId","Surname"], axis=1)
```

```python
features
```

```python
# drop rows missing any data
features = features.dropna()
```

```python
# Use one hot encoding to convert categorical variables into binary format.
features = pd.get_dummies(features, columns=["Geography", "Gender"])
```

```python
features
```

```python
# Goal is for the Data set 80% Training Set and 20% testing set
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.15, random_state=42)
```

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
X_train[0]
```

```python
X_train[1]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pickle #allows saving of models

import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, StackingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
# linear regression
lr_model = LogisticRegression(random_state=42)
```

```python
lr_model.fit(X_train, y_train)
```

```python
# Now that the model is trained it can be used to predict.
```

```python
lr_predictions = lr_model.predict(X_test)
```

```python
lr_predictions
```

```python
lr_accuracy = accuracy_score(y_test, lr_predictions)
```

```python
lr_accuracy
```

```python
# Models: XGBoost, Decision Tree, Random Forest, Naire Bayes, Support Vector Machines, etc.
```

```python
# Function to define models and save them.
def evaluate_and_save_model(model, X_train, X_test,y_train, y_test, filename):
  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)
  accuracy = accuracy_score(y_test, y_pred)
  print(f"{model.__class__.__name__} Accuracy: {accuracy:.4f}")
  print(f"\nClassification Report:\n{classification_report(y_test, y_pred)}")
  print("-----------------------------------")

  with open(filename, "wb") as file:
    pickle.dump(model, file)

    print(f"Model saved as {filename}")
```

```python
xgb_model = xgb.XGBClassifier(random_state=42)
evaluate_and_save_model(xgb_model, X_train, X_test,y_train, y_test, "xgb_model.pkl")

dt_model = DecisionTreeClassifier(random_state=42)
evaluate_and_save_model(dt_model, X_train, X_test, y_train, y_test, "dt_model.pkl")

rt_model = RandomForestClassifier(random_state=42)
evaluate_and_save_model(rt_model, X_train, X_test, y_train, y_test, "rt_model.pkl")

nb_model = GaussianNB()
evaluate_and_save_model(nb_model, X_train, X_test, y_train, y_test, "nb_model.pkl")

knn_model = KNeighborsClassifier()
evaluate_and_save_model(knn_model, X_train, X_test, y_train, y_test, "knn_model.pkl")

svm_model = SVC(random_state=42)
evaluate_and_save_model(svm_model, X_train, X_test, y_train, y_test, "svm_model.pkl")

# GB
gb_model = GradientBoostingClassifier(random_state=42)
evaluate_and_save_model(gb_model, X_train, X_test, y_train, y_test, "gb_model.pkl")

# Stacking
estimators = [
    ('dt', DecisionTreeClassifier(random_state=42)),
    ('rf', RandomForestClassifier(random_state=42)),
    ('svc', SVC(probability=True, random_state=42))
]
stacking_model = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression(),
    cv=5
)
evaluate_and_save_model(stacking_model, X_train, X_test, y_train, y_test, "stacking_model.pkl")
```

```python
# Feature Engineering. Creating the Customer Lifetime Value.
```

```python
features
```

```python
features['CLV'] = df['Balance'] = df['EstimatedSalary'] / 100000
```

```python
features['AgeGroup'] = pd.cut(df['Age'], bins=[0, 30, 45, 60, 100], labels=['Young', 'MiddleAge', 'Senior', 'Elderly'])
```

```python
features
```

```python
features['TenureAgeRatio'] = df['Tenure'] / df['Age']
```

```python
# Handle Categorical Values
features = pd.get_dummies(features, drop_first=True)
```

```python
features
```

```python
# SPlitting into Training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
xgboost_model = xgb.XGBClassifier(random_state=42)
evaluate_and_save_model(xgboost_model, X_train, X_test,y_train, y_test, "xgboost-featureEngineered.pkl")
```

```python
# SMOTE
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=42)
```

```python
X_resampled, y_resampled = smote.fit_resample(X_train,y_train)
```

```python
evaluate_and_save_model(xgboost_model, X_resampled, X_test, y_resampled, y_test, 'xgboost-SMOTE.pkl')
```

```python
# Best model trained so far.
```

```python
# Ensembling using a voting classifier
from sklearn.ensemble import VotingClassifier
```

```python
voting_clf = VotingClassifier(
    estimators = [('xgboost', xgb.XGBClassifier(random_state=42)), ('rf', RandomForestClassifier(random_state=42)), ('svm', SVC(random_state=42, probability=True))],
    voting='hard'
)
```

```python
evaluate_and_save_model(voting_clf, X_resampled, X_test, y_resampled, y_test, 'voting_clf.pkl')
```
